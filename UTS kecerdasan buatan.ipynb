```python
# Sample data for companies and employees
companies = [
    {"name": "Company 1", "country": "United States"},
    {"name": "Company 2", "country": "United Kingdom"},
    {"name": "Company 3", "country": "United States"}
]

employees = [
    {"full_name": "John Doe", "company": "Company 1"},
    {"full_name": "Tom Smith", "company": "Company 2"},
    {"full_name": "Andrew Sebastian", "company": "Company 3"},
    {"full_name": "Ying Han Tan", "company": "Company 1"},
    {"full_name": "Kenneth Ng", "company": "Company 3"},
    {"full_name": "Rubby Lee", "company": "Company 2"},
    {"full_name": "Robert White", "company": "Company 1"}
]

# Function to return employees with company country information
def get_employees_with_country(employees, companies):
    # Create a dictionary to map company names to countries
    company_country_map = {company["name"]: company["country"] for company in companies}
    
    # Create a new list with full employee information
    employee_list = [
        {
            "full_name": employee["full_name"],
            "company": employee["company"],
            "country": company_country_map.get(employee["company"])
        }
        for employee in employees
    ]
    
    return employee_list

# Get the result
result = get_employees_with_country(employees, companies)

# Print the result
for employee in result:
    print(employee)
```

```python
# Sample data for companies and employees
companies = [
    {"name": "Company 1"},
    {"name": "Company 2"},
    {"name": "Company 3"}
]

employees = [
    {"full_name": "John Doe", "company": "Company 1", "nationality": "Australia"},
    {"full_name": "Tom Smith", "company": "Company 2", "nationality": "Hong Kong"},
    {"full_name": "Andrew Sebastian", "company": "Company 3", "nationality": "United States"},
    {"full_name": "Ying Han Tan", "company": "Company 1", "nationality": "Hong Kong"},
    {"full_name": "Kenneth Ng", "company": "Company 3", "nationality": "United States"},
    {"full_name": "Rubby Lee", "company": "Company 2", "nationality": "United States"},
    {"full_name": "Robert White", "company": "Company 1", "nationality": "Australia"}
]

# Function to count employee nationalities for each company
def get_companies_with_employee_nationality(companies, employees):
    # Dictionary to store the count of nationalities for each company
    company_nationality_count = {}

    # Loop through each employee and add to nationality count for their company
    for employee in employees:
        company = employee["company"]
        nationality = employee["nationality"]
        
        # Initialize company entry if it doesn't exist
        if company not in company_nationality_count:
            company_nationality_count[company] = {}

        # Increment nationality count for the employee's company
        if nationality in company_nationality_count[company]:
            company_nationality_count[company][nationality] += 1
        else:
            company_nationality_count[company][nationality] = 1

    # Format the result as a list of dictionaries
    result = [
        {
            "company": company["name"],
            "employee_nationality": company_nationality_count.get(company["name"], {})
        }
        for company in companies
    ]
    
    return result

# Get the result
result = get_companies_with_employee_nationality(companies, employees)

# Print the result
for company_info in result:
    print(company_info)
```
